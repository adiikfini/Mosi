```python
import math
```

```python
g = -9.806
Phi = 3.141592654
ax = 0
ay = g
tstep = 0.01
t = 0
y = 0
x = 0
Vo = float(input("Masukkan Kecepatan Awal (Vo) : "))
sudut = float(input("Masukkan Sudut Kemiringan (sudut) : "))

Vx = Vo * math.cos((sudut*math.pi)/180)
Vy = Vo * math.sin((sudut*math.pi)/180)
```

```python
print("Numerik: ")
while(y >= 0):
    Vy = Vy + (ay*tstep)
    Vx = Vx + (ax * tstep)
    y = y + (Vy * tstep)
    x = x + (Vx * tstep)
    t = t + tstep
    print("y: ", y)
    print("x: ", x)
    print("t: ", t)

```

```python
print("Analitik: ")
x = 0
y = 0
t = 0
Vx = Vo * math.cos((sudut*math.pi)/180)
Vy = Vo * math.sin((sudut*math.pi)/180)
t = 2 * Vy/(-g)
x = x + (Vx*t) + (0.5 * ax * t * t)
y = y + (Vy * t) + (0.5 * ay * t * t)

print("y: ", y, "m")
print("xmax: ", x, "m")
print("tmax: ", t, "s")
```
