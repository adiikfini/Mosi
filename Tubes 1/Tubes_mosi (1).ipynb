# Analitik

# Mengabaikan Hambatan

```python
Vo = 50
sudut = math.radians(35)
g = -9.8
t = 0
y = 0
x = 0

x_isi = [x]
y_isi = [y]
t_isi = [t]

while y >= 0:
  t += 0.01
  t = round(t, 2)
  x = Vo * math.cos(sudut) * t
  y = ((1/2) * g * t**2) + ((Vo * math.sin(sudut)) * t)

  y_isi.append(y)
  x_isi.append(x)
  t_isi.append(t)
```

# Mempertimbangkan Hambatan

```python
m = 0.15
Vo = 50
sudut = math.radians(35)
D = 0.0013
dt = 0.01

g = -9.8
t = 0
vx = Vo * math.cos(sudut)
vy = Vo * math.sin(sudut)
x = 0
y = 0

v = np.sqrt(vx**2 + vy**2)
ax = -(D/m) * v * vx
ay = g - (D/m) *v * vy

list_xya = [[x], [y]]
t_isi = [t]

while y >= 0:
  t += dt
  t = round(t, 2)

  x = (vx * t) + (0.5 * ax * t**2)
  y = (vy * t) + (0.5 * ay * t**2)

  list_xya[0].append(x)
  list_xya[1].append(y)
  t_isi.append(t)
```

```python
Vo = 50
sudut = math.radians(35)
g = -9.8
t = 0
vx = Vo * math.cos(sudut)
vy = Vo * math.sin(sudut)
x = 0
y = 0

xn_isi = [x]
yn_isi = [y]
tn_isi = [t]

while y >= 0:
  t += 0.01
  vy = vy + (g * 0.01)
  x = x + (vx * 0.01)
  y = y + ((vy) * 0.01)
  
  xn_isi.append(x)
  yn_isi.append(y)
  tn_isi.append(t)

mengabaikan = plt.plot(xn_isi, yn_isi)
plt.show()
```

# Mempertimbangkan Hambatan

```python
m = 0.15
Vo = 50
sudut = math.radians(35)
D = 0.0013
dt = 0.01

g = -9.8
t = 0
vx = Vo * math.cos(sudut)
vy = Vo * math.sin(sudut)
x = 0
y = 0

list_xy = [[x], [y]]
list_t = [t]
list_vxy = [[vx], [vy]]


while y >= 0:
  t += dt
  t = round(t, 2)

  v = np.sqrt(vx**2 + vy**2)
  ax = -(D/m) * v * vx
  ay = g - (D/m) *v * vy

  vx = vx + (ax * dt)
  vy = vy + (ay * dt)
  
  x = x + (vx * 0.01)
  y = y + ((vy) * 0.01)
  
  list_xy[0].append(x)
  list_xy[1].append(y)
  list_t.append(t)
  list_vxy[0].append(vx)
  list_vxy[1].append(vy)

mempertimbangkan = plt.plot(list_xy[0], list_xy[1])
plt.show()
```
