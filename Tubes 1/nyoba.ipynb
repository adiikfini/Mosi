```python
import matplotlib.pyplot as plt
import math
import numpy as np
```

Suatu objek dengan massa 0.15 kg ditembakkan dari permukaan tanah dengan 
kecepatan awal 50 m/s dan sudut tembak 35o
. Dengan menggunakan nilai D = 0.0013 
dan  ∆𝑡 = 0.01,
1. Hitung dan bandingkan posisi objek sejak ditembakkan hingga sampai ke
permukaan tanah dengan mengabaikan dan mempertimbangkan hambatan 
udara.

```python
v_awal = 50
sudut = math.radians(35)
g = -9.8
t = 0
y = 0
x = 0

list_xa = [x]
list_ya = [y]
list_ta = [t]

while y >= 0:
  t += 0.01
  t = round(t, 2)
  x = v_awal * math.cos(sudut) * t
  y = ((1/2) * g * t**2) + ((v_awal * math.sin(sudut)) * t)

  list_ya.append(y)
  list_xa.append(x)
  list_ta.append(t)

# t_total = (2 * (v_awal * math.sin(sudut))) / -g
# x_max = v_awal * math.cos(sudut) * t_total
# y_max = ((v_awal**2) * (math.sin(sudut)**2)) / (-2 * g)
```

# Analitik

# Mengabaikan Hambatan

```python
m = 0.15
v_awal = 50
sudut = math.radians(35)
D = 0.0013
dt = 0.01

g = -9.8
t = 0
vx = v_awal * math.cos(sudut)
vy = v_awal * math.sin(sudut)
x = 0
y = 0

v = np.sqrt(vx**2 + vy**2)
ax = -(D/m) * v * vx
ay = g - (D/m) *v * vy

list_xya = [[x], [y]]
list_ta = [t]

while y >= 0:
  t += dt
  t = round(t, 2)

  x = (vx * t) + (0.5 * ax * t**2)
  y = (vy * t) + (0.5 * ay * t**2)

  list_xya[0].append(x)
  list_xya[1].append(y)
  list_ta.append(t)

# t_total = (2 * (v_awal * math.sin(sudut))) / -g
# x_max = v_awal * math.cos(sudut) * t_total
# y_max = ((v_awal**2) * (math.sin(sudut)**2)) / (-2 * g)
```

# Mempertimbangkan Hambatan

```python
v_awal = 50
sudut = math.radians(35)
g = -9.8
t = 0
vx = v_awal * math.cos(sudut)
vy = v_awal * math.sin(sudut)
x = 0
y = 0

list_xn = [x]
list_yn = [y]
list_tn = [t]

while y >= 0:
  t += 0.01
  vy = vy + (g * 0.01)
  x = x + (vx * 0.01)
  y = y + ((vy) * 0.01)
  
  list_xn.append(x)
  list_yn.append(y)
  list_tn.append(t)

mengabaikan = plt.plot(list_xn, list_yn)
plt.show()
```

# Numerik

# Mengabaikan Hambatan

```python
m = 0.15
v_awal = 50
sudut = math.radians(35)
D = 0.0013
dt = 0.01

g = -9.8
t = 0
vx = v_awal * math.cos(sudut)
vy = v_awal * math.sin(sudut)
x = 0
y = 0

list_xy = [[x], [y]]
list_t = [t]
list_vxy = [[vx], [vy]]


while y >= 0:
  t += dt
  t = round(t, 2)

  v = np.sqrt(vx**2 + vy**2)
  ax = -(D/m) * v * vx
  ay = g - (D/m) *v * vy

  vx = vx + (ax * dt)
  vy = vy + (ay * dt)
  
  x = x + (vx * 0.01)
  y = y + ((vy) * 0.01)
  
  list_xy[0].append(x)
  list_xy[1].append(y)
  list_t.append(t)
  list_vxy[0].append(vx)
  list_vxy[1].append(vy)

mempertimbangkan = plt.plot(list_xy[0], list_xy[1])
plt.show()
```
